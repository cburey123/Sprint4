```python
import pandas as pd

df = pd.read_csv("../vehicles_us.csv")
df.head()

#Basic exploratory analysis
df.info()
df.describe()
df.isna().sum()
df.nunique()


#Visualizations with plotly-express
import plotly.express as px

#Distribution of car prices
fig = px.histogram(df, x="price", nbins=100, title="Distribution of Car Prices",
                   labels={"price": "Price (USD)"})
fig.update_xaxes(range=[0, 60000])  # limit extreme outliers
fig.show()

#Price vs Odometer by transmission type
fig = px.scatter(df, x="odometer", y="price", color="transmission",
                 title="Price vs Odometer by Transmission",
                 labels={"odometer": "Odometer (miles)", "price": "Price (USD)"})
fig.update_yaxes(range=[0, 60000])
fig.show()

#Average price by car condition
fig = px.box(df, x="condition", y="price", color="condition",
             title="Car Price Distribution by Condition",
             labels={"condition": "Car Condition", "price": "Price (USD)"})
fig.update_yaxes(range=[0, 60000])
fig.show()

#Days listed distribution
fig = px.histogram(df, x="days_listed", nbins=50, color="condition",
                   title="How Long Cars Stay Listed by Condition",
                   labels={"days_listed": "Days Listed"})
fig.show()

#Paint color by popularity
fig = px.histogram(df, x="paint_color", color="paint_color",
                   title="Distribution of Paint Colors",
                   labels={"paint_color": "Paint Color"})
fig.show()

```
